```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### `univariate analysis`

### x1

```python
sns.distplot(data['x1'])
```

```python
sns.boxplot(data['x1'])
```

```python
data['x1'].describe()
```

```python
data['x1'].skew()
```

```python
q1=np.percentile(data['x1'],25)
q3=np.percentile(data['x1'],75)
IQR=q3-q1
good_value=1.5 * IQR
lower_bound=q1-good_value
upper_bound=q3+good_value
print(f'Lower Bound = {lower_bound} and Upper Bound = {upper_bound}')
```

```python
data = data[data['x1'] <= upper_bound]   # remove outliers
```

### x2

```python
sns.distplot(data['x2'])
```

```python
sns.boxplot(data['x2'])
```

```python
q1=np.percentile(data['x2'],25)
q3=np.percentile(data['x2'],75)
IQR=q3-q1
good_value=1.5 * IQR
lower_bound=q1-good_value
upper_bound=q3+good_value
print(f'Lower Bound = {lower_bound} and Upper Bound = {upper_bound}')
```

```python
data = data[data['x2'] <= upper_bound]   # remove outliers
```

```python
data['x2'].describe()
```

```python
data['x2'].skew()
```

```python
sns.distplot(data['x2'])
```

### y

```python
data['y'].value_counts()
```

```python
sns.countplot(data['y'])
```

### `Bivariate analysis`

```python
cont_columns = ['x1','x2']
```

```python
from scipy.stats import ttest_ind

for i in cont_columns:
    target_yes = data.loc[data['y']==1,i]
    target_no = data.loc[data['y']==0,i]
    ttest_results = ttest_ind(target_yes,target_no)
    print(f'{i} vs y => {ttest_results}')
```

```python
X = data[['x1','x2']]
y = data['y']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train[['x1','x2']])
```

```python
X_train[['x1', 'x2']] = sc.transform(X_train[['x1','x2']])
X_test[['x1','x2']] = sc.transform(X_test[['x1','x2']])
```

## `Logistic Regression ---------------`

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_proba_logistic = model.predict_proba(X_test)
y_pred_proba_logistic
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score,recall_score,f1_score
print(f'Confusion matrix =>\n {confusion_matrix(y_test,y_pred)}')
print(f'Accuracy => {accuracy_score(y_test,y_pred)}')
print(f'Precision => {precision_score(y_test,y_pred)}')
print(f'Recall => {recall_score(y_test,y_pred)}')
print(f'f1-score => {f1_score(y_test,y_pred)}')
```

##### Here in this hackathon I have tested 5 models :
    Logistic Regression
    Decision Tree
    KNN
    SVM with Linear Kernel
    SVM with RBF Kernel
    
   

```python
from sklearn.metrics import roc_curve,roc_auc_score
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))

ns_prob = [0 for _ in range(len(y_test))]  # bad algo
ns_fpr,ns_tpr,thresholds = roc_curve(y_test.astype(int),ns_prob)

lr_prob = y_pred_proba_logistic[:,1]
lr_auc = roc_auc_score(y_test,lr_prob)   # logistic regression
lr_fpr,lr_tpr,thesholds = roc_curve(y_test,lr_prob)

dt_prob = y_pred_proba_tree[:,1]
dt_auc = roc_auc_score(y_test,dt_prob)   # Decision tree
dt_fpr,dt_tpr,thesholds = roc_curve(y_test,dt_prob)

knn_prob = y_pred_proba_knn[:,1]
knn_auc = roc_auc_score(y_test,knn_prob)   # knn
knn_fpr,knn_tpr,thesholds = roc_curve(y_test,knn_prob)

svcl_prob = y_pred_proba_svclin[:,1]
svcl_auc = roc_auc_score(y_test,svcl_prob)   # svc linear
svcl_fpr,svcl_tpr,thesholds = roc_curve(y_test,svcl_prob)

svcrfb_prob = y_pred_proba_scvrfb[:,1]
svcrfb_auc = roc_auc_score(y_test,svcrfb_prob)   # svc rfb
svcrfb_fpr,svcrfb_tpr,thesholds = roc_curve(y_test,svcrfb_prob)


# plotting
plt.plot(lr_fpr,lr_tpr,marker='.',label='Logistic Regression')
plt.plot(dt_fpr,dt_tpr,marker='.',label='Decision tree')
plt.plot(knn_fpr,knn_tpr,marker='.',label='KNN')
plt.plot(svcl_fpr,svcl_tpr,marker='.',label='SVC Linear Kernel')
plt.plot(svcrfb_fpr,svcrfb_tpr,marker='.',label='SVC RFB Kernel')
plt.plot(ns_fpr,ns_tpr,marker='.',label='Bad Random Algo')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()


print(f'Logistic Regression AUC is => {lr_auc}')
print(f'Decision Tree AUC is => {dt_auc}')
print(f'Knn AUC is => {knn_auc}')
print(f'SVC Linear Kernel AUC is => {svcl_auc}')
print(f'SVC RFB Kernel AUC is => {svcrfb_auc}')
```
